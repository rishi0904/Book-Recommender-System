This part of the code is responsible for building the item-item similarity matrix
We first convert build a list of dictionary.
By creating groups of books coreeEach dictionary corresponds to a single book. The user_id is the key, while rating given by the user for the book is its value.

```python
listOfDictonaries=[]
#for item-item based similarity matrix this will be used
indexMap = {}
reverseIndexMap = {}
#to keep tract of the index
ptr=0;
testdf = ratingsDF
#we make a testdf to not disturb the ratings matrix
testdf=testdf[['user_id','rating']].groupby(testdf['book_id'])
#this will group all the books together so that there will be users and their ratings for each group(book)
#we are using the groupby method to group users and ratings according to books they have rated
for groupKey in testdf.groups.keys():
    #testdf.group.keys() will give all the groups formed by the groupby method and groupKey will take every group one by one
    tempDict={}
    #temp dict for temporary storing data in a dictionary to add to listofDictionaries
    groupDF = testdf.get_group(groupKey)
    #get_group method will fetch all the data corresponding to the groupKey and store it in groupDf
    for i in range(0,len(groupDF)):
        #loop through all the entries of the group to users of a book
        tempDict[groupDF.iloc[i,0]]=groupDF.iloc[i,1]
        #making a dict by assigning the user the rating it has given
        #each tempdict corresponds to a single book
    indexMap[ptr]=groupKey
    #keeping track of the index
    reverseIndexMap[groupKey] = ptr
    ptr=ptr+1
    listOfDictonaries.append(tempDict)
    #adding the temp dict of a book to listOfDictionaries with all the other dicts of books
```

We then use sklearn's DictVectorizer() function to create vectors corresponding to each book. We are trying to create a vector space with users as column vectors. Each point in the vector space represents a book. Rating of the book given an user is its magnitude. We then calculate similarity/distance between books in this vector space.

```python
from sklearn.feature_extraction import DictVectorizer
dictVectorizer = DictVectorizer(sparse=True)
#sparse is True means the matrix will not have NaN values instead it'll have zeros in place.
vector = dictVectorizer.fit_transform(listOfDictonaries)
```

We finally use sklearn's consine_similarity function to calculate pairwise similarity matrix 

```python
from sklearn.metrics.pairwise import cosine_similarity
pairwiseSimilarity = cosine_similarity(vector)
#define: A.B=||A||.||B||.cos@
#where cos@ is the simiarity
```

At last we define a function that takes an input id and outputs 5 top most recommendations. Please note that this id is the column id in book.csv. Values of id are in  [1,10000]

```python
def printBookDetails(bookID):
    print("Title:", bookDF[bookDF['id']==bookID]['original_title'].values[0])
    print("Author:",bookDF[bookDF['id']==bookID]['authors'].values[0])
    print("Printing Book-ID:",bookID)
    print("=================++++++++++++++=========================")


def getTopRecommandations(bookID):
    row = reverseIndexMap[bookID]
    print("------INPUT BOOK--------")
    printBookDetails(bookID)
    print("-------RECOMMENDATIONS----------")
    similarBookIDs = [printBookDetails(indexMap[i]) for i in np.argsort(pairwiseSimilarity[row])[-7:-2][::-1]]
    print(similarBookIDs)
```

```python
def getTopRecommandationsname(bookname):
    
    bookID=(bookDF[bookDF['original_title']==bookname]['id'].values[0])
    row = reverseIndexMap[bookID]
    #reverse indexing of the simialarity row of that book in the pairwise similarity matrix 
    print("------INPUT BOOK--------")
    printBookDetails(bookID)
    print("-------RECOMMENDATIONS----------")
    #print((np.argsort(pairwiseSimilarity[row])[-7:-2][::-1]))
    #this is a np.array of length 10000 
    for i in np.argsort(pairwiseSimilarity[row])[-7:-2][::-1]:
        print(i)
        printBookDetails(indexMap[i])
```
